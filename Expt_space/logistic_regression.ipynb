```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X, Y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=Y)
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

We can conclude the following from the general equation of a straight line  
w0 + w1x + w2y = 0:
The slope is given by -w1/w2, given that w2 ≠ 0. The x-intercept is given by -w0/w1 and the y-intercept is given by -w0/w2.

```python
def perceptron(X,Y):
  X=np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  lr=0.1

  for i in range(1000):
    j = np.random.randint(0,100) # slecting a random number out of the 100 samples we created
    P = sigmoid(np.dot(X[j],weights))
    weights = weights+lr*(Y[j]-P)*X[j]

  return weights[0], weights[1:]
```

```python
intercept, coef = perceptron(X,Y)
```

```python
m= -(coef[0]/coef[1])
b= -(intercept/coef[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X,Y)
LogisticRegression()
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
```

```python
plt.figure(figsize=(15,8))
plt.plot(x_input,y_input, color='red')# perceptorn wala
plt.plot(x_input1, y_input1, color = 'black')# logistic regression wala
plt.scatter(x=X[:,0],y=X[:,1],c=Y)
plt.ylim(-3,2)
plt.show()
```
